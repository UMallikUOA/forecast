```python
import pandas as pd
import numpy as np
import re

#data loading
pth = "forecast_history.xlsx"
data = pd.read_excel(pth)
data
```

```python
# cleaning the data
def cleanData(data):
    #as the data has non numeric values we will be removing those
    data = data.replace({'\$': '', 'I': '1','%':''}, regex=True)
    data = data.apply(pd.to_numeric, errors ='coerce')
    data.fillna(0,inplace=True)
    return data
```

```python
cols = ['Westpac: 4 year forecast', 'Joe Bloggs: 2 year forecast','Harry Spent: 5 year forecast']

df=cleanData(data)
df
```

```python
#forecast calculation
df['Westpac'] = df['Median house price'] * (1+df['Westpac: 4 year forecast']/100)
df['Joe Bloggs'] = df['Median house price'] * (1+df['Joe Bloggs: 2 year forecast']/100)
df['Harry Spent'] = df['Median house price'] * (1+df['Harry Spent: 5 year forecast']/100)
df
```

```python
# function to calculate MAPE metric
def met_mape(actual_val, forecasted):
    return np.mean(np.abs((actual_val-forecasted)/actual_val)) * 100
```

```python
westpac_mape = met_mape(df['Median house price'],df['Westpac: 4 year forecast'])
joe_mape=met_mape(df['Median house price'],df['Joe Bloggs: 2 year forecast'])
harry_mape=met_mape(df['Median house price'],df['Harry Spent: 5 year forecast'])
```

```python
print(f'Westpac: {westpac_mape}')
print(f'Joe Bloggs: {joe_mape}')
print(f'Harry Spent: {harry_mape}')
```
